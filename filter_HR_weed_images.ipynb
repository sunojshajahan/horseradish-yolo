```python
import os
import shutil

# source dataset
root = "Object-Detection-Horseradish-#2-2"   # folder with train/val/test
# output folder
out_root = "filtered_dataset"

# class ids
horseradish_id = 0
weed_id = 1

splits = ["train", "valid", "test"]

for split in splits:

    src_img = os.path.join(root, split, "images")
    src_lbl = os.path.join(root, split, "labels")

    dst_img = os.path.join(out_root, split, "images")
    dst_lbl = os.path.join(out_root, split, "labels")

    os.makedirs(dst_img, exist_ok=True)
    os.makedirs(dst_lbl, exist_ok=True)

    for fname in os.listdir(src_lbl):
        if not fname.endswith(".txt"):
            continue

        lbl_path = os.path.join(src_lbl, fname)

        # read label file
        with open(lbl_path, "r") as f:
            lines = f.read().strip().splitlines()

        if not lines:
            continue

        # class ids present
        classes = [int(line.split()[0]) for line in lines]

        # must contain both
        if horseradish_id in classes and weed_id in classes:

            # copy label
            shutil.copy(lbl_path, os.path.join(dst_lbl, fname))

            base = os.path.splitext(fname)[0]

            # image extension guess
            found = False
            for ext in ("jpg", "jpeg", "png"):
                img_path = os.path.join(src_img, f"{base}.{ext}")
                if os.path.exists(img_path):
                    shutil.copy(img_path, os.path.join(dst_img, f"{base}.{ext}"))
                    found = True
                    break

            if not found:
                print("image missing for", fname)

print("done")
```
